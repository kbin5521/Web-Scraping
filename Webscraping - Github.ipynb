```python
import requests
import pygal
from pygal.style import LightColorizedStyle, LightenStyle
import pandas as pd
## 打印出这个API 调用的基本信息
url = 'https://api.github.com/search/repositories?q=language:python&sort=stars'
r = requests.get(url)
print('Status code:', r.status_code)

response_dict = r.json() #把获得的response存成 json 文档

print(response_dict.keys())

repo_dicts = response_dict['items']
print('Repositories returned:', len(repo_dicts))

repo_dict_0 = repo_dicts[0]
print('\nKeys in Items:', len(repo_dict_0))
# for key in sorted(repo_dict_0.keys()):
#     print(key)


# print('\nSelected information about first repository:')
# print('NAME:', repo_dict_0['name'])
# print('Owner:', repo_dict_0['owner']['login'])
# print('Stars:', repo_dict_0['stargazers_count'])
# print('REPOSITORY:', repo_dict_0['html_url'])
# print('Description:', repo_dict_0['description'])
# print('Home page:', repo_dict_0['homepage'])



print('\nSelected information about first repository:')
for repo_dict in repo_dicts[0:3]:
    print('\nNAME:', repo_dict['name'])
    print('Owner:', repo_dict['owner']['login'])
    print('Stars:', repo_dict['stargazers_count'])
    print('REPOSITORY:', repo_dict['html_url'])
    print('Description:', repo_dict['description'])
    print('Home page:', repo_dict['homepage'])
```

```python
# 将调出来的信息汇成 svg 可视化图片. 
names, liss = [], []

for repo_dict in repo_dicts:
    names.append(repo_dict['name'])
    
    lis = {
        'value': repo_dict['stargazers_count'],
        'label': str(repo_dict['description']),
        'xlink': repo_dict['html_url'],
    }
    liss.append(lis)


        
my_config = pygal.Config()
my_config.x_label_rotation = 60
my_config.show_legend = False
my_config.title_font_size = 100
my_config.label_font_size = 50
my_config.major_label_font_size = 38
my_config.truncate_label = 35
my_config.show_y_guides
my_config.width = 1000
            
my_style = LightenStyle('#333366',base_style = LightColorizedStyle)
chart = pygal.Bar(my_config, style = my_style)
chart.title = 'Most-Starred Python projects on GitHub'
chart.x_labels = names

chart.add('',liss)
chart.render_to_file('python_repos.svg')

```
